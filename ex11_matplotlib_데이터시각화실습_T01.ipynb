# Bar chart

```python
np.random.seed(6)
bar_x=np.arange(1,4)
bar_y=np.random.randint(50,100,size=3)
plt.bar(bar_x,bar_y)
plt.show()
```

```python
plt.bar(bar_x,bar_y,color=['limegreen','mediumslateblue','orange'])
for i in range(len(bar_x)):
    plt.text(bar_x[i],bar_y[i],f'{bar_y[i]}',
             fontdict={'color':'black','size':14})
plt.xticks(range(1,4))
plt.show()
```

## 예제2

```python
economi=pd.read_csv('./data/시도_성별_경제활동인구_총괄_KOSIS_2022.csv',
                   encoding='euc-kr')
economi.head()
```

```python
economi.info()
```

```python
economi.isnull().sum()
```

```python
economi['고용률(%)'].head()
```

```python
emp_pt=economi['취업자 (천명)']/economi['15세이상인구 (천명)']
emp_pt=round(emp_pt,3)*100
economi['고용률(%)']=emp_pt
economi['고용률(%)'].head()
```

```python
economi['실업률(%)'].tail()
```

```python
unemp_pt=economi['실업자 (천명)']/economi['경제활동인구 (천명)']
unemp_pt=round(unemp_pt,3)*100
economi['실업률(%)']=unemp_pt
economi['실업률(%)'].tail()
```

```python
economi.head()
```

```python
state_emp=economi[['행정구역(시도)','고용률(%)']].groupby('행정구역(시도)',as_index=False).mean()
state_emp
```

```python
plt.figure(figsize=(8,4))
plt.bar(economi['행정구역(시도)'],economi['고용률(%)'])
plt.xlabel('행정구역(시도)')
plt.ylabel('고용률(%)')
plt.title('시도별 고용률 평균 시각화')
plt.xticks(rotation=80)
plt.ylim(60,80)
plt.show()
```

```python
plt.figure(figsize=(8,4))
plt.barh(economi['행정구역(시도)'],economi['고용률(%)'])
plt.xlim(50,80)
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(x=state_emp['행정구역(시도)'],y=state_emp['고용률(%)'],palette='husl')
plt.xticks(rotation=85)
plt.show()
```

```python
economi.head()
```

```python
man=economi.query('성별=="남자"')
man_unemp=man[['행정구역(시도)','실업률(%)']]
man_unemp.head()
```

```python
# warning 없애기
import warnings
warnings.filterwarnings(action='ignore')
```

# Scatter plot, Pie chart

```python
np.random.seed(2)
x=np.arange(1,4)
y=np.random.rand(3)
print(x,y)

plt.scatter(x,y)
plt.show()
```

```python
for i in y:
    print(f'{i/y.sum()*100:.2f}')
    
plt.pie(y,labels=x,autopct='%.1f%%',explode=[0.05,0.05,0.1],shadow=True,counterclock=False,startangle=130)
plt.show()
```

# 종합 최종예제

```python
acc_data=pd.read_csv('./data/Traffic_Accident_2017.csv',
                    encoding='euc-kr')
acc_data.head(2)
```

```python
display(acc_data.shape,
       acc_data.info())
```

```python
acc_data.isnull().sum()
```

```python
tmp=acc_data['요일'].value_counts()
tmp=tmp[['월','화','수','목','금','토','일']]
tmp
```

```python
plt.bar(tmp.index,tmp)
plt.title('요일별 교통사고 건수')
plt.ylabel('사고건수')
plt.xlabel('요일')
plt.show()
```

```python
acc_data.info()
```

```python
acc_data1=acc_data[['발생지시도','사고유형_대분류','사망자수','사상자수','중상자수','경상자수']]
acc_data1
```

```python
result=acc_data.query('사고유형_대분류=="차대차"')[['발생지시도','사상자수','사망자수','중상자수','경상자수']].groupby('발생지시도').sum().sort_values(by='사망자수',ascending=False)
result
```

```python
plt.figure(figsize=(10,4))
sns.barplot(result.index,result['사망자수'])
plt.title('발생지 시도별 차대차 사건 사상자 수')
plt.xlabel('발생지 시도')
plt.ylabel('사망자 수')
plt.savefig('data/요일별_교통사고_건수.png') # 이미지 저장 
plt.show()
```

```python
acc_data.info()
```

```python
acc_data['발생시간']=acc_data['발생년월일시']%100
acc_data.head(2)
```

```python
result=acc_data.groupby('발생시간').agg(cnt=('발생시간','count'))
result
```

```python
result=acc_data['발생년월일시']%100
result=result.value_counts().sort_index()
result
```

```python
sns.set_palette('pastel')
explode=[0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
plt.pie(result,labels=result.index,autopct='%.1f%%',shadow=True,explode=explode,counterclock=False,startangle=130)
plt.savefig('./data/발생일시별.png')
plt.show()
```

```python
bins=[-1,2,5,8,11,14,17,20,23]
labels=['0~2','3~5','6~8','9~11','12~14','15~17','18~20','21~23']
resultCut=pd.cut(acc_data['발생년월일시']%100,bins=bins,labels=labels)
resultCut=resultCut.value_counts()
resultCut
```

```python
colors=['gold','silver','lightgray','whitesmoke','whitesmoke','whitesmoke','whitesmoke','whitesmoke']
explode=[0.15,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
plt.pie(resultCut,labels=resultCut.index,autopct='%1.0f%%',colors=colors,explode=explode,startangle=250,counterclock=False,shadow=True)
plt.savefig('./data/교통사고가 가장 많은 시간대별.png')
plt.show()
```

```python
acc_data.info()
```

```python
result=acc_data.query('발생지시도=="광주"')['법규위반'].value_counts()
result
```

```python
colors=['gold','silver','lightgray','whitesmoke','whitesmoke','whitesmoke','whitesmoke','whitesmoke','whitesmoke','whitesmoke']
explode=[0.15,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05,0.05]
plt.pie(result,labels=result.index,autopct='%1.0f%%',colors=colors,explode=explode,startangle=350,counterclock=False,shadow=True)
plt.savefig('./data/지역선정 법규위반 사항별.png')
plt.xticks(rotation=85)
plt.show()
```
