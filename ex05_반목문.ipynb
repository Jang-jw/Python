### while 문
- 반복횟수가 명확하지 않을 때

```python
# num 이라는 변수를 선언하고
# 파이썬 최고!!! 13 번 출력해보기
num=0
while cnt<13:
    num+=1
    print(f'{cnt} : 파이썬 최고!!!')
```

```python
# break
while True:
    print('출력')
    break
```

```python
# continue
# 프로그램을 실행하다가 continue 문장을 만나면 continue 아래 문장은 무시가 되고
# 처음 조건을 판별하는 코드로 되돌아 감 (while num < 10 :)
num=0
while num<10:
    num+=1 # num = num+1
    if num%2==0:
        continue
    print(num)
```

```python
# 두개의 정수를 입력 받아서 더하는 코드를 작성하시오.
while True:
    num1=int(input('첫 번째 정수 입력 >> '))
    num2=int(input('두 번째 정수 입력 >> '))
    if num1==0 and num2==0:
        print('프로그램이 종료되었습니다.')
        break
    else:
        print(f'두 정수의 합 : {num1+num2}')
```

```python
nowWeight=int(input('현재 몸무개: '))
targetWeight=int(input('목표 몸무개: '))
cnt=0
while True:
    cnt+=1
    val=int(input(f'{cnt}주차 감량 몸무게: '))
    nowWeight-=val
    if nowWeight<=targetWeight:
        print(f'{nowWeight}kg 달성! 축하합니다!!')
        break
```

### for 문
- 반복횟수가 정해져 있을 때

```python
lunch=['큐브 식판밥','김치찌개','한솥','돈까스']
lunch
```

```python
for i in lunch:
    print(i)
```

```python
for i in ['큐브 식판밥','김치찌개','한솥','돈까스']:
    print(i)
```

```python
for i in '큐브 식판밥':
    print(i)
```

```python
len(lunch)
```

```python
hi = '안녕하세요!'
for i in hi:
    print(i,end='') # print() 속성 : 기본값은 개행해서 출력
                    # shift + tab 으로 함수 속성 확인
                    # 출력문의 끝 속성 : end = '' or "" 
```

```python
score_list=[90,45,70,60,55]
for i in range(len(score_list)):
    if score_list[i]>=60:
        print(f'{i+1}번 학생은 합격입니다.')
    else:
        print(f'{i+1}번 학생은 불합격입니다.')
        
print()

# 풀이
for i in score_list:
    if i>=60:
        print(f'{score_list.index(i)+1}번 학생은 합격입니다.')
    else:
        print(f'{score_list.index(i)+1}번 학생은 불합격입니다.')
```

#### for : range()
- 필요한 만큼의 숫자를 만들어 내는 유용한 기능
- range(시작할 숫자, 종료할 숫자, 증감량)

```python
lunch
```

```python
for i in range(1,len(lunch)):
    print(i,lunch[i])
```

```python
# 풀이 
# range 함수를 사용하여 합격/불합격 여부 판별하기
# range 함수의 시작값의 기본값 => 0, 증감값의 기본값 => 1
for i in range(len(score_list)): # 0 ~ 리스트의 길이만큼 숫자 반환 
    # print(i)
    if score_list[i]>=60:
        print(f'{i+1}번 학생은 합격입니다.')
    else:
        print(f'{i+1}번 학생은 불합격입니다.')
```

```python
for i in range(1,25,3):
    print(i,end=' ')
```

### min, max 함수
- 선언한 자료구조(리스트,튜플)에서 가장 큰 값, 가장 작은 값을 찾아주는 함수

```python
list2=[4,5,2,9,12,3,7,28,69,45,100]
# list2 에서 가장 작은 수 구하기
min(list2)
# list2 에서 가장 큰 수 구하기
max(list2)
```

```python
sum=0
for i in range(3,100,3):
    sum+=i
print(sum)
```
