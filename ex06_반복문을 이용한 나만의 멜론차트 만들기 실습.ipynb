### melon top_100 데이터를 활용한 나만의 멜로차트 만들기

```python
import pickle
```

```python
# <Pickle> 
# - 파이썬 리스트, 튜플 등의 자료구조 형태를 유지하면서
# 파일로 저장, 불러올 때 사용하는 모듈
# - 규칙이 있는 데이터의 형식을 유지하면서 데이터 저장
# - 바이너리 형태로 저장 (저장용량이 다른 파일에 비해 작다)
# - 불러올때 : rb(read binary)
# - 저장할때 : wb (write binary)

# <melon data>
# [차트순위, 노래제목, 가수이름]
```

```python
# 멜론차트의 데이터를 불러와서 데이터 확인하기
# with 문 사용해서 데이터 불러오기
with open('music_100.pkl','rb') as f: # 'rb' : read binary 
    music=pickle.load(f)
```

```python
music[:5]
```

```python
s_song=input('제목 중 찾고 싶은 키워드를 입력하세요 >> ')
print(f'검색한 노래가 위치한 index : ')
result=[]
for i in music:
    if s_song in i[1]:
        # print(i)
        result.append([i[1],i[2]])
        print(music.index(i),end=' ')
print()
print(f'\'{s_song}\'으로 검색한 노래 수 : {len(result)}')
result
```

```python
# 특정 가수의 노래 리스트 만들기

# 1. 검색할 가수명 입력받기

# 2. for 반목문을 이용하여 전체 데이터를 가져오기

# 3. for 반목변수가 꺼내온 정보 확인 (인덱스, 데이터 위치, 데이터 형태)
# 내가 입력받은 가수명이 꺼내온 정보의 어디에 있는지 확인

# 4. 데이터가 있다면 노래 리스트에 담아주기

# 5. 노래 리스트 출력하기

result=[]
singer=input('가수명을 입력해 주세요 >> ')
for i in range(len(music)):
    if music[i][2]==singer:
        result.append(music[i])
#result.sort()
result
```

```python
result=[]
cnt=0
singer='임영웅'
for i in range(len(music)):
    if music[i][2]==singer:
        music[i][2]='장재욱'
        result.append(music[i])
        cnt+=1
        #result.sort()
print(f'melon 차트 속 나의 곡 : {cnt}곡')
result
```

```python
# 내가 만든 나만의 멜론차트를 파일로 저장하기
with open('mymelon.pkl','wb') as f:
    pickle.dump(music,f)
```

### 네이버 영화리뷰 데이터를 활용한 분석

```python
# 데이터 불러와서 확인하기
with open('movie_review.pkl','rb') as f:
    m=pickle.load(f)
```

```python
m[:15]
# 인덱스 0 : 텍스트 리뷰정보, 1 : 긍정/부정 감성 리뷰정보
```

```python
# 전체 데이터 갯수

len(m)
```

```python
# 0번 텍스트 리뷰의 길이
len(m[0][0])
```

```python
# 전체 리뷰의 길이를 구하기
cnt=0
for i in m:
    cnt+=len(i[0])
cnt
```

```python
# 전체 리뷰길이 평균값을 구해보자
cnt/len(m)
```

```python
# 긍정 / 부정 리뷰의 개수를 구해보자
# 긍정 / 부정 리뷰의 데이터에 접근 (인덱싱)
cntP=0
cntN=0
for i in m:
    if i[1]==1:
# 긍정
        cntP+=1
#부정
    else:
        cntN+=1
print(f'긍정개수:{cntP}, 부정개수:{cntN}')
```

```python
# 리뷰길이의 최대 / 최소값을 확인하고 그 내용을 출력하기
cntMax=0
cntMin=0
indexMax=0
indexMin=9999
# 1. min, max 함수를 사용하여 최대 / 최소값 확인하기
for i in range(len(m)):
    l=len(m[i][0])
    if l>cntMax:  # 0
        cntMax=len(m[i][0]) # 19
        indexMax=i   # 0
    elif l<cntMin:
        cntMin=len(m[i][0])
        indexMin=i
# 2. 최대 길이의 리뷰 봅아서 출력하기 - 개수도 함께 확인
print(f'최대 길이의 리뷰:{m[indexMax][0]}, 리뷰길이:{cntMax}')
# 3. 최소 길이의 리뷰 뽑아서 출력하기 - 개수도 함깨 확인
print(f'최소 길이의 리뷰:{m[indexMin][0]}, 리뷰길이:{cntMin}')
```

```python
lenList=[]
for i in range(len(m)):
    lenList.append(len(m[i][0]))
# 2. 최대 길이의 리뷰 봅아서 출력하기 - 개수도 함께 확인
print(f'최대 길이의 리뷰:{m[lenList.index(max(lenList))][0]}, 리뷰길이:{max(lenList)}')
# 3. 최소 길이의 리뷰 뽑아서 출력하기 - 개수도 함깨 확인
print(f'최소 길이의 리뷰:{m[lenList.index(min(lenList))][0]}, 리뷰길이:{min(lenList)}')
```
