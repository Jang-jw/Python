### 예제 4

```python
# 예제 4
```

```python
center=np.array([0,0])
tree=np.array([[1,0],[2,0],[3,0]])
radius=2
display(center,tree,radius)
```

```python
for i in range(len(tree)):
    distance=np.sqrt((tree[i,0]-center[0])**2+
                     (tree[i,1]-center[1]**2))
    if distance==2:
        print(f'{i+1} 번 나무의 거리는 {distance} 으로 \'on\' 입니다.')
    elif distance<2:
        print(f'{i+1} 번 나무의 거리는 {distance} 으로 \'in\' 입니다.')
    elif distance>2:
        print(f'{i+1} 번 나무의 거리는 {distance} 으로 \'out\' 입니다.')
    else:
        continue
```

# 최종 예제

```python
data=np.loadtxt('data/ratings.dat',delimiter='::',dtype=np.int64)
data
```

```python
display(data.ndim,data.shape,data.size)
```

```python
data[data[:,0]==1]
```

```python
dataLen=np.unique(data[:,0])
dataLen
```

```python
dataLen.size
```

```python
for i in dataLen:
    j=data[data[:,0]==i,2].mean()
    print(f'{i} 번째 사용자의 평점 평균은 {j} 점 입니다.')
```

```python
result=[]
for i in dataLen:
    j=data[data[:,0]==i,2].mean()
    if j>=4:
        result.append([i,j])
    else:
        continue
result
```

```python
len(result)
```

```python
print(f'평점 평균이 4점 이상인 사용자는 {len(result)} 명 입니다.')
```
