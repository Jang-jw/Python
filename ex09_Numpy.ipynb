#### 조건에 맞는 데이터 접근하기
- 불리언 인덱싱

```python
# random.choice : 랜덤으로 하나의 요소를 취득
# random.sample : 랜덤으로 여러 요소를 취득, 중복 없음
# random.choices : 랜덤으로 여러 요소를 취득, 중복 있음
```

```python
# 불리언 배열 생성
sample_arr=[True,False]
np.random.seed(1)
bool_arr=np.random.choice(sample_arr,size=10)
display(bool_arr)
tmp_data=np.arange(10) # 0 ~ 9 까지 1 씩 증가하는 숫자 생성
display(tmp_data)
```

```python
bool_arr
```

```python
tmp_data
```

```python
tmp_data[bool_arr]
```

#### sqrt() : 제곱근 함수

- 제곱근(square root)

```python
square_arr=np.array([4,16])
print(square_arr)
# print(square_arr.sqrt()) 에러 : 메소드로 정의가 되어있지 않음
print(np.sqrt(square_arr))
```

```python
# sqrt() 함수를 이용
tree=(3,0) # B(X2,Y2)
tree_x,tree_y=tree[0],tree[1]
distance=np.sqrt((tree_x-0)**2+(tree_y-0)**2)

if distance==2.0:
    print('on')
elif distance<2.0:
    print('in')
elif distance>2.0:
    print('out')
```

#### 전체 평점 평균 구하기

```python
data
```

```python
data[:,2].mean()
```

```python
data
```

```python
# 1. 사용자 아이디만 접근
dataID=data[:,0]
dataID
```

```python
dataID.size
```

```python
# 2. 사용자 아이디를 중복없이 가져오기
dataIDU=np.unique(dataID)
dataIDU
```

```python
dataIDU.size
```

```python
# 3. 사용자 아이디가 1인 평점의 평균 구하기
data1=data[data[:,0]==1]
data1
```

```python
data1[:,2].mean()
```

```python
for i in dataIDU:
    print(f'{i}번째 사용자의 평점 평균 : {data[data[:,0]==i][:,2].mean()}')
```

```python
result=[]
for i in dataIDU:
    j=data[data[:,0]==i][:,2].mean()
    if j>=4:
        result.append(j)
result
```

```python
len(result)
```

```python
print(f'평점이 4점 이상인 사용자 수 : {len(result)}')
```

```python
resultList=[]
for i in dataIDU:
    resultList.append([i,data[data[:,0]==i][:,2].mean()])
resultList
```

```python
# csv 파일로 저장 
np.savetxt('./data/user_rating_mean.csv',resultList,delimiter=',',fmt='%.3f')
```
