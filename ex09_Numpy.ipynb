#### sqrt() :

```python
square_arr=np.array([4,16])
print(square_arr)
# print(square_arr.sqrt())  :    
print(np.sqrt(square_arr))
```

```python
# sqrt()  
tree=(3,0) # B(X2,Y2)
tree_x,tree_y=tree[0],tree[1]
distance=np.sqrt((tree_x-0)**2+(tree_y-0)**2)

if distance==2.0:
    print('on')
elif distance<2.0:
    print('in')
elif distance>2.0:
    print('out')
```
