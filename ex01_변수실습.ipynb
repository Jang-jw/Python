```python
# find vs index 차이점 확인하기 : 문장안에서 없는 문자열 찾을 때
day.find('sunny') # 찾는 문자열이 없을 경우 -1 반환, 에러를 발생하지 않음
# 신뢰성 있는 코드를 위해 지향
# day.index('sunny') 
# 문자를 찾을 수 없다면 에러 발생
```

```python
# replace ('찾을 문자열', '변경할 문자열') : 특정 문자열을 찾아서 변경하는 함수
# day : 맑음 => 눈
day=day.replace('맑음','눈')
day
```

```python
# strip() : 양족 공백을 제거해 주는 함수
day2='     오늘은 즐거운 금요일입니다.      '
day2
```

```python
print(day2.lstrip()) # 왼쪽 공백을 제거하는 함수
print(day2.rstrip()) # 오른쪽 공백을 제거하는 함수
```

### 문자열 포매팅

- format 함수를 사용한 포매팅

#### %기호 포매팅!

```python
month=6
day=22
s='오늘은 %d월 %d일 입니다.' %(month,day)
s
```

```python
# 3개이상 값을 포매팅할때
year=2023
month=11
day=10
s='오늘은 우리반 수료하는 날이에요 - %d년 %d월 %d일 ^^' %(year,month,day)
s
```

```python
# 포매팅 실습 예제 1
lang='"python"'
s='Life is too short, You need %s'%lang
print(s)
s=f'Life is too short, You need {lang}'
print(s)
```
