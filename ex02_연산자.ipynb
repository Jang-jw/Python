### 산술 연산자

#### 숫자 연산

```python
num1=10
num2=7

# 나누기 /, 나머지 %, 몫 //
print(num1/num2)
print(num1%num2)
print(num1//num2)
```

#### 문자열 더하기

```python
str1='안녕'
str2='하세요!'

# 문자열 +
str1+str2
```

```python
str1='10'
str2='7'

# 숫자를 문자열로 + 
str1+str2

# 문자열 체크하는 함수 : isdigit(), isalpha()

# 문자열이 문자인지 아닌지를 확인하는 함수 : isalpha()
# 숫자가 '단 하나'라도 있다면 False 반환
# 한글도 문자로 인식 => True

# 문자열이 숫자인지 아닌지를 확인하는 함수 : isdigit()
# 문자가 '단 하나'라도 있다면 False 반환
# 음수부호 -, 소수점 . => 숫자가 아닌 문자로 판단, 실수나 음수를 판단하지 못함
# 오직 0 을 포함한 양수형 정수, 지수 => True

str1.isdigit()

# 문자열 전처리(내가 분석하고자 하는 문제에 맞게 깔끔하게 데이터를 정리하는 작업)
# => 조건문 활용
```

#### 숫자, 문자열을 더했을 때

```python
num1=10 # 정수 자료형
str1 = '7' # 문자열 자료형

# 형변환 : (int)x => int(대상)
# 숫자가 연산이 되도록 형변환
print(num1+int(str1))

# 문자가 연산이 되도록 형변환
print(str(num1)+str1)
```

```python
num1=int(input('정수를 입력하세요 >> '))
num2=int(input('정수를 입력하세요 >> '))
print('더하기 결과 : ',num1+num2)
print('빼기 결과 : ',num1-num2)  
print('곱하기 결과 : ',num1*num2)
print('나누기 결과 : ',num1/num2)
```

```python
print('********** 연산하기 **********')
num1=int(input('정수입력 >> '))
num2=int(input('정수입력 >> '))
cal=input('+ 연산자 입력 >> ')
'{} {} {} = {}'.format(num1,cal,num2,num1+num2)
```

### 비교연산자
- True/False 로 결과가 출력되는 연산자  
 : >, >=, <, <=, ==(같다), !=(같지않다)

```python
a=3
b=7
print(a>b)
print(a==b)
print(a!=b)
print(a<b)
```

### 논리연산자
- not : 논리를 뒤집는 연산자 True -> False => False => True
- and, or

```python
# not
a,b=5,10
not a < b # a < b : True => not => False
```

### 멤버 연산자

```python
# not in
'우리반' not in str1
```

### 삼항 연산자
- True 실행문장 if 조건식 else False 실행문장

```python
# 60점 이상이면 합격, 미만이면 불합격 출력하기
score=70
'합격' if score >= 60 else '불합격'
```

```python
a=int(input('정수입력 >> '))
b=int(input('정수입력 >> '))
print('a :',a) if a>b else print('b :',b)
f'a : {a}' if a>b else f'b : {b}'
```

```python
# 삼항 연산자 실습 예제 3
# 키보드로 정수를 입력 받아 홀수인지 짝수인지 판별하시오.

num=int(input('정수 입력 >> '))
f'{num}는(은) 홀수입니다.' if num%2==1 else f'{num}는(은) 짝수입니다.'
# print(num,'는(은) 홀수입니다.') if num%2==1 else print(num,'는(은) 짝수입니다.')
```
