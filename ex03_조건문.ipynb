```python
cnt_mask=int(input('   >> '))
a=8
b=5
result8=0
result5=0
while(True):
    if(cnt_mask//a>0):
        result8+=1
        cnt_mask-=a
    elif(cnt_mask//b>0):
        result5+=1
        cnt_mask-=b
    elif(cnt_mask%b>0):
        result5+=1
        cnt_mask=0
    else:
        break
print(f'{a}   : ',result8)
print(f'{b}   : ',result5)
```

```python
#  score        .
score=int(input('  >> '))
if(score>=90):
    grade='A'
elif(score>=80):
    grade='B'
elif(score>=70):
    grade='C'
elif(score>=60):
    grade='D'
else:
    grade='F'
print(f'{score} {grade} .')
```
